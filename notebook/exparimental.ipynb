```python
# importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load the dataset
df = pd.read_csv('../BTCUSDT_1d_full_klines.csv')
```

```python
df.head()
```

```python
# here we have to drop the follow columns becuse we don't need it
df.drop(columns={'timestamp','close_time'},inplace=True)
```

```python
# lets check the correlation of the columns with the close column
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
plt.title('Correlation Metrix of the Columns')
plt.show()
```
